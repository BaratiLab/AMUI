# Datasets

```python
import matplotlib.pyplot as plt
import numpy as np

from datasets import load_dataset
```

```python
ds = load_dataset(
    "baratilab/Eagar-Tsai",
    "process_maps",
    split="m_Ti64_p_0_480_20_v_0.0_2.9_0.1",
    streaming="true",
)
```

```python
print(ds)
print(ds.features)
image = ds.features["lengths"]
print(image)
```

```python
for data in ds:
  for key, value in data.items():
    print(key, np.array(value).shape)

plt.imshow(data["depths"])
plt.gca().invert_yaxis()
plt.colorbar()
plt.show()
plt.imshow(data["lengths"])
plt.gca().invert_yaxis()
plt.colorbar()
plt.show()
plt.imshow(data["widths"])
plt.gca().invert_yaxis()
plt.colorbar()
plt.show()
```

```python
for data in ds:
  for key, value in data.items():
    if (key in ["width", "depth", "length"]):
      print(value)
    print(key, np.array(value).shape)
```
