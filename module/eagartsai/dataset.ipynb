# Datasets

```python
import matplotlib.pyplot as plt
import numpy as np

from datasets import load_dataset
```

```python
ds = load_dataset("baratilab/Eagar-Tsai", streaming="true", split="Ti64")
```

```python
print(ds)
print(ds.features)
image = ds.features["depths"]
print(image)
```

```python
for data in ds:
  for key, value in data.items():
    print(key, np.array(value).shape)

plt.imshow(data["depths"])
plt.gca().invert_yaxis()
plt.colorbar()
plt.show()
plt.imshow(data["lengths"])
plt.gca().invert_yaxis()
plt.colorbar()
plt.show()
plt.imshow(data["widths"])
plt.gca().invert_yaxis()
plt.colorbar()
plt.show()
```
