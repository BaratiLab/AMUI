# Datasets

```python
import matplotlib.pyplot as plt
import numpy as np

from datasets import load_dataset
```

```python
ds = load_dataset(
    "baratilab/Eagar-Tsai",
    "process_maps",
    split="m_IN625_p_0_480_20_v_0.0_2.9_0.1",
    streaming="true",
)
```

```python
print(ds)
print(ds.features)
```

```python
for data in ds:
  for key, value in data.items():
    print(key, np.array(value).shape)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

img1 = axs[0, 0].imshow(np.array(data["depths"]) * 1E6)
axs[0, 0].invert_yaxis()
axs[0, 0].set_title("Depths")
axs[0, 0].set_xlabel("Velocity (mm/s)")
axs[0, 0].set_xticks([])
axs[0, 0].set_ylabel("Power (W)")
axs[0, 0].set_yticks([])
cbar1 = fig.colorbar(img1, ax=axs[0, 0])
cbar1.set_label("μm")

img2 = axs[0, 1].imshow(np.array(data["lengths"]) * 1E3)
axs[0, 1].invert_yaxis()
axs[0, 1].set_title("Lengths")
axs[0, 1].set_xlabel("Velocity (mm/s)")
axs[0, 1].set_xticks([])
axs[0, 1].set_ylabel("Power (W)")
axs[0, 1].set_yticks([])
cbar2 = fig.colorbar(img2, ax=axs[0, 1])
cbar2.set_label("mm")

img3 = axs[1, 0].imshow(np.array(data["widths"]) * 1E6)
axs[1, 0].invert_yaxis()
axs[1, 0].set_title("Widths")
axs[1, 0].set_xlabel("Velocity (mm/s)")
axs[1, 0].set_xticks([])
axs[1, 0].set_ylabel("Power (W)")
axs[1, 0].set_yticks([])
cbar3 = fig.colorbar(img3, ax=axs[1, 0])
cbar3.set_label("μm")

img4 = axs[1, 1].imshow(np.array(data["widths"])/np.array(data["depths"]), vmax=5)
axs[1, 1].invert_yaxis()
axs[1, 1].set_title("Widths / Depths")
axs[1, 1].set_xlabel("Velocity (mm/s)")
axs[1, 1].set_xticks([])
axs[1, 1].set_ylabel("Power (W)")
axs[1, 1].set_yticks([])
fig.colorbar(img4, ax=axs[1, 1])

plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)
plt.suptitle("Eagar-Tsai Simulations for IN625", fontsize=18)
plt.show()
```
